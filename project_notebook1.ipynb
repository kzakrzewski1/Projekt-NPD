```python
import pandas as pd
import numpy as np
import os

pop = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_4751604.csv', skiprows = 4)
pop = pop.dropna(how='all', axis='columns')

gdp = pd.read_csv('API_NY.GDP.MKTP.CD_DS2_en_csv_v2_4751562.csv', skiprows = 4)
gdp = gdp.dropna(how='all', axis='columns')

co2 = pd.read_csv('fossil-fuel-co2-emissions-by-nation_csv.csv')


pop['Country Name'] = pop['Country Name'].str.upper()
gdp['Country Name'] = gdp['Country Name'].str.upper()
```

```python
co2 = co2[co2['Year'].isin(range(start, finish + 1))]

co2
```

```python
co2_ = pd.merge(co2, countries, on = 'Country')

co2_ = co2_.groupby(['Year', 'Country Code'], as_index = False).sum(numeric_only = True)
co2_ = pd.merge(co2_, gdp[['Country Name', 'Country Code']], on = 'Country Code').sort_values(['Year', 'Country Code'])
co2_.drop('Per Capita', axis =  1, inplace = True)

co2_
```

```python
def get_population(year, country_code):

    if  pop[pop['Country Code'] == country_code].loc[:, str(year)].values:
        return pop[pop['Country Code'] == country_code].loc[:, str(year)].values[0]
    else:
        return np.nan

def get_gdp(year, country_code):

    if  gdp[gdp['Country Code'] == country_code].loc[:, str(year)].values:
        return gdp[gdp['Country Code'] == country_code].loc[:, str(year)].values[0]
    else:
        print(country_code)
        return np.nan


def per_capita(total, population):
    return 1000 * total/population

def GDP_per_capita(gdp, population):
    return gdp/population




co2_['Population'] = co2_[['Year', 'Country Code']].apply(lambda x:
                                                          get_population((x['Year']), x['Country Code']), axis = 1)

co2_['GDP'] = co2_[['Year', 'Country Code']].apply(lambda x:
                                                   get_gdp((x['Year']), x['Country Code']), axis = 1)

co2_["Per Capita"] = co2_[['Total', 'Population']].apply(lambda x:
                                                         per_capita(x['Total'], x['Population']), axis = 1)

co2_['GDP Per Capita'] = co2_[['GDP', 'Population']].apply(lambda x:
                                                             GDP_per_capita(x['GDP'], x['Population']), axis = 1)


co2_ = co2_[co2_['Per Capita'].notna()]

```

```python
def n_highest_emissions_pc(data, n = 5):
    return data.sort_values(by = ['Year', 'Per Capita'],
                                 ascending = [True, False]).groupby(['Year']).head(n)[['Year',
                                                                                      'Country Name', 
                                                                                      'Per Capita', 
                                                                                      'Total']]

def n_highest_emissions_pc_year(data, year, n = 5):
    try:
        assert (year >= start and year <= finish)
        return data[data['Year'] == year].sort_values(by =  'Per Capita',
                                                                ascending = False).head(n)[['Year', 
                                                                                            'Country Name', 
                                                                                            'Per Capita', 
                                                                                            'Total']]
    except AssertionError:
        print('No data available for this year')

n_highest_emissions_pc_year(co2_, 1970)


```

```python
def n_highest_gdps_pc(data, n = 5):
    return data.sort_values(by = ['Year', 'GDP Per Capita'],
                                 ascending = [True, False]).groupby(['Year']).head(n)[['Year',
                                                                                      'Country Name', 
                                                                                      'Total', 
                                                                                      'GDP Per Capita']]

def n_highest_gdps_pc_year(data, year, n = 5):
    try:
        assert (year >= start and year <= finish)
        return data[data['Year'] == year].sort_values(by =  'GDP Per Capita',
                                                                ascending = False).head(n)[['Year', 
                                                                                            'Country Name', 
                                                                                            'GDP Per Capita', 
                                                                                            'Total']]
    except AssertionError:
        print('No data available for this year')

n_highest_gdps_pc_year(co2_, 1970, 3)
```

```python
def get_reduction(country_name, start, end):
    if  (co2_[(co2_['Country Name'] == country_name) & (co2_['Year'] == start)]['Per Capita'].values and
         co2_[(co2_['Country Name'] == country_name) & (co2_['Year'] == end)]['Per Capita'].values): 
        return (co2_[(co2_['Country Name'] == country_name) & (co2_['Year'] == start)]['Per Capita'].values[0] 
              - co2_[(co2_['Country Name'] == country_name) & (co2_['Year'] == end)]['Per Capita'].values[0])
    else:
        return np.nan
    

def n_highest_reduction(start, finish, year, n):
    try:
        assert (year > start and year <= finish)
        if (year - 10 < start):
                print(f"Available data goes back only for {year - start} years. Following table shows countries with highest reduction in the last {year - start} years.")

        begin = max(year - 10, start)
        country_list['Reduction'] = country_list[['Country Name']].apply(lambda x: get_reduction(x['Country Name'], begin, year), axis = 1)

        return country_list.sort_values(by = 'Reduction', ascending = False).dropna().head(n)
        
    except AssertionError:
        if (start == finish):
            print("The available data comes from only one year")
        else:
            print(f"Required data is not available for this year. Please select a year between {start+1} and {finish}")


country_list =  pd.DataFrame(co2_["Country Name"].unique())
country_list.columns = ['Country Name']

n_highest_reduction(start, finish, 2014, 5)
```
